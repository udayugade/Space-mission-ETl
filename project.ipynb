```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df['Price'].fillna(0, inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
price_data = df.dropna(subset=['Price', 'Year'])
```

```python
plt.figure(figsize=(10,6))
plt.hist(price_data['Price'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Launch Prices')
plt.xlabel('Price (in million USD)')
plt.ylabel('Number of Launches')
plt.show()
```

```python
launches_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.plot(launches_per_year.index, launches_per_year.values, marker='o')
plt.title('Number of Launches Over the Years')
plt.xlabel('Year')
plt.ylabel('Launch Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(price_data['Year'], price_data['Price'], alpha=0.7, color='purple')
plt.title('Launch Price Over the Years')
plt.xlabel('Year')
plt.ylabel('Price (in million USD)')
plt.show()
```

```python
org_count = df['Organisation'].value_counts().head(10)
plt.figure(figsize=(10,6))
org_count.plot(kind='bar', color='skyblue')
plt.title('Top 10 Organisations by Number of Launches')
plt.xlabel('Organisation')
plt.ylabel('Number of Launches')
plt.xticks(rotation=45)
plt.show()
```

```python
status_count = df['Mission_Status'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(status_count, labels=status_count.index, autopct='%1.1f%%', startangle=90)
plt.title('Mission Success Rate')
plt.show()
```
